### Notebook 11: Hypothesis testing

In this notebook we illustrate the need to demonstrate the validity of any assertions we may wish to make as part of any statistical analysis. For example, suppose you are a venture capitalist and have received the proposition of a project which has been running for two years. Based on the data you have at hand, the decision to make is whether or not you should invest in this project. Suppose you will decide to invest if average monthly profit is at least Â£7,000. This situation does not require you to explicitly estimate some parameter values. Instead, you need to make a judgement on whether or not the condition is satisfied. For such situations we need the statistic tool that is hypothesis testing.


````python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
# import microsoft.csv, and add a new feature - logReturn

ms = pd.read_csv('data/microsoft.csv')
ms['logReturn'] = np.log(ms['Close'].shift(-1)) - np.log(ms['Close'])
````

````python
# log return goes up and down during the period

ms['logReturn'].plot(figsize=(16, 6))
plt.axhline(0, color='red')
plt.show()
````

### Steps involved in testing a claim by hypothesis testing

#### Step 1: Set hypothesis

$H_0 : \mu = 0$ 
$H_a : \mu \neq 0$

$H_0$ means the average stock return is equal to 0 <br>
$H_a$ means the average stock return is not equal to 0

#### Step 2: Evaluate the test statistic

````python
sample_mean = ms['logReturn'].mean()
sample_std = ms['logReturn'].std(ddof=1)
n = ms['logReturn'].shape[0]

# when the sample size is large enough (as is the case here), 
# we can use the z-distribution instead of the t-distribtuion 
# [also mu = 0 under the null hypothesis]

zhat = (sample_mean - 0)/(sample_std/(n**0.5))
print(zhat)
````

#### Step 3: Set the desicion criteria

````python
# confidence level
alpha = 0.05

zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft  # z-distribution is symmetric 
print(zleft, zright)
````

#### Step 4:  Decision time - shall we reject $H_0$?

````python
print('At significant level of {}, shall we reject: {}'
      .format(alpha, zhat>zright or zhat<zleft))
````
